```python
import numpy as np
import pandas as pd
import csv
import glob

import matplotlib.pyplot as plt

from scipy.stats import kstest
from scipy.stats import normaltest
from scipy.stats import norm

from itertools import combinations
import pyvinecopulib as pv

import seaborn as sns
from scipy import stats
```

```python
# Assuming you have three different CSV files with file paths
file1 = r'CompassPaper/HKNW.csv'
file2 = r'CompassPaper/FACT.csv'
file3 = r'CompassPaper/HRYR.csv'

# Read the three CSV files into separate dataframes
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)
df3 = pd.read_csv(file3)

# Get the column names
NBO_dat = df1.dropna()
CT_dat = df2.dropna()
KGL_dat = df3.dropna()
```

```python
datasets = [NBO_dat, CT_dat, KGL_dat]  # List of datasets

def normal_test(series):
    series = pd.to_numeric(series, errors='coerce')
    series = series.dropna()
    if len(series) < 8:
        return float('NaN')  # or any value indicating insufficient samples
    statistic, p_value = normaltest(series)
    return p_value

# Iterate over the datasets
for i, dataset in enumerate(datasets):
    print(f"Dataset {i+1}:")
    # Apply the normaltest to each column of the DataFrame
    p_values = dataset.apply(normal_test)
    
    # Print the results
    print(p_values)

    # Check the p-value against a significance level (e.g., 0.05)
    for column, p_value in p_values.items():
        if pd.isna(p_value):
            print(f"The variable {column} in dataset {i+1} has insufficient samples for the test")
        elif p_value < 0.05:
            print(f"The variable {column} in dataset {i+1} does not come from a Normal distribution")
        else:
            print(f"The variable {column} in dataset {i+1} comes from a Normal distribution")
    print()
```

```python
datasets = [NBO_dat, CT_dat, KGL_dat]

dataset_names = ['Nairobi Wilson Airport (HKNW)', 'Cape Town International Airport (FACT)', 
                 'Kigali International Airport (HRYR)']  # List of dataset names

# Iterate over the datasets
for i, dataset in enumerate(datasets):
    print(f"{dataset_names[i]}:")
    
    # Exclude the first column
    dataset = dataset.iloc[:, 1:]
    
    # Create subplots for each variable
    fig, axs = plt.subplots(nrows=2, ncols=4, figsize=(12, 6))
    fig.suptitle(f"{dataset_names[i]}", fontsize=14)
    axs = axs.flatten()
    
    # Iterate over the columns
    for j, column in enumerate(dataset.columns):
        # Plot histogram and KDE for each column
        sns.histplot(dataset[column], ax=axs[j], bins=20, kde=True)
        axs[j].set_title(f"{column}")
        axs[j].set_xlabel("Values")
        axs[j].set_ylabel("Density")
    
    # Adjust layout and spacing
    fig.tight_layout()
    plt.subplots_adjust(top=0.85)
    
    plt.show()
    print()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(42)

# Generate random data
n_samples = 1000
t_max_24 = np.random.uniform(20, 50, n_samples)
t_min_24 = np.random.uniform(5, 35, n_samples)
copula_params = 2.90

# Compute copula values
u = np.random.uniform(size=n_samples)
v = np.random.uniform(size=n_samples)
copula = -1 / copula_params * np.log((np.exp(-copula_params * u) - 1) / (np.exp(-copula_params * v) - 1))

# Scatter plot
plt.scatter(t_max_24, t_min_24, c=copula, cmap='coolwarm')

# Axes labels
plt.xlabel('t_max_24')
plt.ylabel('t_min_24')

# Colorbar
cbar = plt.colorbar()
cbar.set_label('Copula')

# Title
plt.title(f'Scatter Plot - Copula: Frank, Parameters: {copula_params}')

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(42)

# Generate random data
n_samples = 1000
t_max_24 = np.random.uniform(20, 50, n_samples)
t_min_24 = np.random.uniform(5, 35, n_samples)
copula_params = 1.09

# Compute copula values
u = np.random.uniform(size=n_samples)
v = np.random.uniform(size=n_samples)
copula = np.exp(-((-np.log(u)) ** copula_params + (-np.log(v)) ** copula_params) ** (1 / copula_params))

# Scatter plot
plt.scatter(t_max_24, t_min_24, c=copula, cmap='coolwarm')

# Axes labels
plt.xlabel('t_max_24')
plt.ylabel('t_min_24')

# Colorbar
cbar = plt.colorbar()
cbar.set_label('Copula')

# Title
plt.title(f'Scatter Plot - Copula: Gumbel, Parameters: {copula_params}')

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pyvinecopulib as pv

# List of copulas to test
copulas = [pv.Bicop(pv.BicopFamily.gaussian), pv.Bicop(pv.BicopFamily.student),
           pv.Bicop(pv.BicopFamily.clayton), pv.Bicop(pv.BicopFamily.gumbel),
           pv.Bicop(pv.BicopFamily.frank)]

u = np.linspace(0.01, 0.99, 100)
v = np.linspace(0.01, 0.99, 100)
U, V = np.meshgrid(u, v)

for copula in copulas:
    copula.fit(np.column_stack((U.flatten(), V.flatten())))

    Z = copula.pdf(U, V).reshape(U.shape)

    fig = plt.figure()
    ax = fig.add_subplot(111)
    contour = ax.contour(U, V, Z)
    ax.set_xlabel('U')
    ax.set_ylabel('V')
    ax.set_title(f'{copula.name} Copula Contour Plot')
    plt.colorbar(contour)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal, t
from copulas.multivariate import GaussianMultivariate
from copulas.univariate import Clayton, Gumbel, Frank

# List of copulas to test
copulas = [GaussianMultivariate, t, Clayton, Gumbel, Frank]
copula_names = ["Gaussian", "Student's t", "Clayton", "Gumbel", "Frank"]

u = np.linspace(0.01, 0.99, 100)
v = np.linspace(0.01, 0.99, 100)
U, V = np.meshgrid(u, v)

for copula, copula_name in zip(copulas, copula_names):
    copula_instance = copula()
    copula_instance.fit(np.column_stack((U.flatten(), V.flatten())))

    Z = copula_instance.pdf(U, V).reshape(U.shape)

    fig = plt.figure()
    ax = fig.add_subplot(111)
    contour = ax.contour(U, V, Z)
    ax.set_xlabel('U')
    ax.set_ylabel('V')
    ax.set_title(f'{copula_name} Copula Contour Plot')
    plt.colorbar(contour)
    plt.show()
```
